```python
import pandas as pd
import numpy as np
```

```python
df_visit = pd.read_csv('./cdm_data/visit_occurrence.csv')
```

```python
df_con = df_con[['person_id','condition_concept_id','condition_start_datetime','condition_end_datetime']]
df_con.head(2)
```

```python
df_drug = pd.read_csv('./cdm_data/drug_exposure.csv')
df_drug.head(2)
```

```python
df_drug = df_drug[['person_id','drug_concept_id','drug_exposure_start_datetime','drug_exposure_end_datetime','visit_occurrence_id']]
df_drug.head(2)
```

```python
df_cct = df_cct[['concept_id','concept_name','domain_id']]
df_cct.head(2)
```

```python
df_cct.columns=['condition_concept_id','concept_name','domain_id']
df_cct.head(2)
```

```python
merge_1 = pd.merge(df_visit,df_per, how='outer', on='person_id')
merge_1.head(2)
```

```python
merge_2 = pd.merge(merge_1,df_con, how='outer', on='person_id')
merge_2.head(2)
```

```python
merge_3 = pd.merge(merge_2,df_drug, how='outer', on='visit_occurrence_id')
merge_3.head(2)
```

```python
df_visit_per = pd.merge(merge_3,df_cct, how='left', left_on='condition_concept_id', right_on='condition_concept_id')
df_visit_per.head(2)
```

```python
df_visit_per = df_visit_per[df_visit_per.visit_concept_id == 9201]
df_visit_per.head(2)
```

```python
len(df_visit_per)
```

```python
import datetime as dt
```

```python
from datetime import datetime
```

```python
df_visit_per_a = df_visit_per
df_visit_per.info()
```

```python
df_visit_per_a['visit_start_datetime'] = df_visit_per['visit_start_datetime'].apply(lambda x:x[:10])
df_visit_per_a.head(2)
```

```python
df_visit_per_a['visit_start_datetime'] = pd.to_datetime(df_visit_per_a['visit_start_datetime'], format='%Y-%m-%d', errors='raise')
df_visit_per_a.head(2)
```

```python
df_visit_per_a['visit_end_datetime'] = df_visit_per_a['visit_end_datetime'].apply(lambda x:x[:10])
df_visit_per_a['visit_end_datetime'] = pd.to_datetime(df_visit_per_a['visit_end_datetime'], format='%Y-%m-%d', errors='raise')
df_visit_per_a.head(2)
```

```python
df_visit_per_a['birth_datetime'] = df_visit_per_a['birth_datetime'].apply(lambda x:x[:10])
df_visit_per_a['birth_datetime'] = pd.to_datetime(df_visit_per_a['birth_datetime'], format='%Y-%m-%d', errors='raise')
df_visit_per_a.head(2)
```

```python
df_visit_per_b = df_visit_per_a.sort_values(['person_id_x','visit_end_datetime'], ascending=True)
```

```python
df_visit_per_b.head(30)
```

```python
p_li = [0]
for i in range(len(df_visit_per_b['person_id_x'])-1):
    if df_visit_per_b.iloc[i,1] == df_visit_per_b.iloc[i+1,1] and df_visit_per_b.iloc[i,10] != df_visit_per_b.iloc[i+1,10]:
        t = df_visit_per_b.iloc[i+1,3] - df_visit_per_b.iloc[i,4]
        t = str(t).split(' ')[0]
        p_li.append(t)
    else:
        p_li.append(0)
        
p_li
```

```python
len(p_li)
```

```python
len(df_visit_per_b)
```

```python
df_visit_per_b['re_visit'] = p_li
df_visit_per_b.head(2)
```

```python
df_visit_per_b['re_visit'] = df_visit_per_b['re_visit'].astype(int)
```

```python
pp_li = []
for i in df_visit_per_b['re_visit']:
    if 0 < i <= 30:
        pp_li.append(1)
    else:
        pp_li.append(0)
        
pp_li
```

```python
count = 0
for i in pp_li:
    if i == 1:
        count = count+1
        
count
```

```python
df_visit_per_b['re_visit_m'] = pp_li
df_visit_per_b.head(2)
```

```python
df_visit_per_b.drop('person_id_y', axis=1, inplace=True)
```

```python
df_visit_per_b.rename(columns={'person_id_x':'person_id'}, inplace=True)
```

```python
df_visit_per_b.drop('condition_end_datetime', axis=1, inplace=True)
```

```python
df_visit_per_c = df_visit_per_b.fillna(0)
```

```python
df_visit_per_c.isna().sum()
```

```python
df_visit_per_c.head(2)
```

```python
y = df_visit_per_c['re_visit_m']
x = df_visit_per_c.drop(['re_visit_m','birth_datetime','gender_source_value','ethnicity_concept_id',
                        'condition_start_datetime','drug_exposure_start_datetime','drug_exposure_start_datetime',
                        'drug_exposure_end_datetime','domain_id','visit_start_datetime','visit_end_datetime','re_visit'], axis=1)
```

```python
x
```

```python
x = pd.get_dummies(x)
x.info()
```

```python
x.head(2)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=111)
```

```python
# from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.dummy import DummyClassifier
from sklearn.metrics import classification_report
```

```python
tree = DecisionTreeClassifier()
knn = KNeighborsClassifier()
svc = SVC()
# dummy = dummy()
```

```python
tree.fit(x_train, y_train)
knn.fit(x_train, y_train)
svc.fit(x_train, y_train)
# dummy_majority = DummyClassifier(strategy='most_frequent').fit(x_train, y_train)
```

```python
tree_preds = tree.predict(x_test)
knn_preds = knn.predict(x_test)
svc_preds = svc.predict(x_test)
# dummy_preds = dummy.predict(x_test)
```

```python
print(classification_report(y_test, tree_preds))
print('='*100)
print(classification_report(y_test, knn_preds))
print('='*100)
print(classification_report(y_test, svc_preds))
# print('='*100)
# print(classification_report(y_test, dummy_preds))
```
