```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
concept = pd.read_csv('./cdm_data/concept.csv')
concept
```

```python
person_origin = pd.read_csv('./cdm_data/person.csv')
visit_origin = pd.read_csv('./cdm_data/visit_occurrence.csv')
drug_origin = pd.read_csv('./cdm_data/drug_exposure.csv')
condition_origin = pd.read_csv('./cdm_data/condition_occurrence.csv')
```

# Question
* 입원했던 환자가 퇴원 후 30일 이내 재입원할 확률
* 모델 사용 시점 - 환자 퇴원시점

```python
# 데이터 프레임 압축 및 merge
```

```python
visit_origin.info()
```

```python
visit = visit_origin.iloc[:,[0,1,2,3,5]]
```

```python
person_origin.info()
```

```python
person = person_origin.iloc[:,[0,1,5,6]]
```

```python
drug_origin.info()
```

```python
# drug -> one_hot_vector
drug_ohv = pd.concat([
    drug_origin.iloc[:,17], 
    pd.get_dummies(drug_origin.drug_concept_id,prefix='drug',prefix_sep='_')
],axis=1)
drug_ohv
```

```python
# one-hot vector -> visit 마다의 약 처방
drug = drug_ohv.groupby('visit_occurrence_id').sum()
drug
```

```python
condition_origin.info()
```

```python
condition_ohv = pd.concat([condition_origin.visit_occurrence_id, pd.get_dummies(condition_origin.condition_concept_id,prefix='condi',prefix_sep='_')],axis=1)
```

```python
# visit 별 진단받은 진단 총 계
condition = condition_ohv.groupby('visit_occurrence_id').sum()
```

```python
visit[['visit_start_date','visit_end_date']] = visit[['visit_start_date','visit_end_date']].astype('datetime64')
```

```python
# drug랑 condition visit에 맞춰 merge
comb_tmp = pd.merge(visit,condition,how='left',on='visit_occurrence_id')
comb = pd.merge(comb_tmp,drug,how='left',on='visit_occurrence_id') 
```

```python
# 사람으로 정렬하면서 사람끼리 모으고 거기에 visit start date로 정렬해 사람마다 시간순 정렬 적용
comb.sort_values(['person_id','visit_start_date'],inplace=True)
```

```python
# nan값은 condition 이나 drug 가 없는 경우이므로 그냥 0으로 채워줌
comb.fillna(0, inplace=True)
comb
```

```python
# 각 사람마다 진단이나 약물 처방의 누적횟수를 생성
comb[comb.iloc[:,5:].columns.tolist()] = comb.groupby('person_id')[comb.iloc[:,5:].columns.tolist()].transform(np.cumsum)
```

```python
# 단순하게 방문 당시의 진단이나 약물 처방을 추가했어야했는데 깜빡한듯?
# 오우 각 사람 입원마다 입원해 있던 시간을 추가 및 입원 누적 시간 추가하는것도 좋을듯 
comb
```

```python
# 입원만 추려내서
combIP = comb.query('visit_concept_id == 9201')
```

```python
concept.query('concept_id in [9201, 9202, 9203]')
```

```python
# 입원 횟수가 1이면 사전에 추려내고
ipcnt = combIP.groupby('person_id').size()
ipperson = ipcnt[ipcnt>1].index.tolist()
```

```python
target_check = combIP.query(f'person_id in {ipperson}')
```

```python
target_check.reset_index(drop=True, inplace = True)
target_check
```

```python
# 재입원에 소요된 일자 추가
arr = []
for row in range(len(target_check)-1):
    if target_check.loc[row,'person_id'] != target_check.loc[row+1,'person_id']:
        arr.append(np.nan)
    else:
        arr.append(target_check.loc[row+1,'visit_start_date'] - target_check.loc[row,'visit_end_date'])
arr.append(np.nan)
target_check['revisit'] = arr
        
```

```python
target_check
```

```python
# 마지막 방문은 nan값이니 제거
target = target_check.dropna()
```

```python
# target 항목을 추가해 30일 이내 재입원했으면 1, 아니라면 0의 라벨을 부여
target['target'] = target['revisit'].apply(lambda date: 1 if date <= np.timedelta64(30,'D') else 0) 
```

```python
# 각 사람마다 정보 추가
target = pd.merge(target, person, on='person_id',how='inner')
target
```

```python
# 나이 항목 만들어서 계산 후 추가
target['age'] = ((target['visit_end_date']-target['birth_datetime'].astype('datetime64'))/np.timedelta64(1,'Y')).astype(int)
target
```

```python
# 필요없으니 지움
target.drop(['revisit','birth_datetime'],axis=1,inplace=True)
```

```python
target.columns
```

```python
# 컨디션과 drug 중 모든 person_id에 대해서 0값을 가지는 건 싹다 삭제
target = target.drop(target.sum()[target.sum()==0].index.tolist(),axis=1)
```

```python
# 전처리 끝
target
```

```python
# trial
y = target['target']
x = target.drop(['visit_occurrence_id','person_id','visit_concept_id','visit_start_date','visit_end_date','target'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=11)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
xgb = XGBClassifier(
    objective='binary:logistic',
    gpu_id=0,
    tree_method='gpu_hist',
    eta=0.1,
    n_estimators=100,
    min_child_weight=10,
    gamma=0,
    max_depth=5,
    learning_rate=0.01,
    subsample = 0.9
)
xgb.fit(x_train,y_train)
pred_train = xgb.predict_proba(x_train)
pred = xgb.predict_proba(x_test)
pred1 = xgb.predict(x_test)
print(roc_auc_score(y_test, pred[:,1]), roc_auc_score(y_train, pred_train[:,1]))
```
