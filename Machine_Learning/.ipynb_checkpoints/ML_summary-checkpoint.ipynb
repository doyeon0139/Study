```python
df = pd.read_csv('./ml_source_files/loan_data.csv')
y = df['not.fully.paid']
x = df.drop(['not.fully.paid','purpose'], axis=1)
```

```python
x = pd.get_dummies(x)
```

```python
x.head(2)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
```

```python
tree = DecisionTreeClassifier()
lr = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC(probability=True)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
models = [tree, lr, knn, svm]
```

```python
vot_h = VotingClassifier(estimators = [('tree', tree),('lr', lr),('knn',knn),('svm', svm)], voting='hard')
vot_s = VotingClassifier(estimators = [('tree', tree),('lr', lr),('knn',knn),('svm', svm)], voting='soft')
```

```python
for i, j in enumerate(models):
    j.fit(x_train, y_train)
    accuracy = j.score(x_test, y_test)
    
    print(j.__class__.__name__,':', accuracy)
    print('='*100)
```
