```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv('./ml_source_files/loan_data.csv')
```

```python
df.head(3)
```

```python
y = df['not.fully.paid']
x = df.drop(['not.fully.paid','purpose'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
```

```python
tree = DecisionTreeClassifier()
lr = LogisticRegression()
knn = KNeighborsClassifier()
gnb = GaussianNB()
svc = SVC(probability=True)
```

```python
vot_h = VotingClassifier(estimators=[('lr', lr),('tree', tree),('knn', knn),('gnb', gnb),('svc', svc)], voting='hard')
vot_s = VotingClassifier(estimators=[('lr', lr),('tree', tree),('knn', knn),('gnb', gnb),('svc', svc)], voting='soft')
```

```python
models = [tree, lr, knn, gnb, svc, vot_h, vot_s]
```

```python
for i,j in enumerate(models):
    j.fit(x_train, y_train)
    s = j.score(x_test, y_test)
    print(j.__class__.__name__, ':', s)
    print('========================================')
```
