```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./deep_data/housing.csv', header=None, delim_whitespace=True)
df.head()
```

```python
df.info()
```

```python
y = df.values[:,13]
x = df.values[:,:13]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=100)
```

```python
df.values.shape
```

```python
x_train
```

```python
model = Sequential()
model.add(Dense(30, activation='relu', input_dim=13))
model.add(Dense(20, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
model.summary()
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
hist = model.fit(x_train, y_train, validation_split=0.2, epochs=200, batch_size=20)
```

```python
model.evaluate(x_test, y_test)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_sc = sc.transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
hist = model.fit(x_train_sc, y_train, validation_split=0.2, epochs=200, batch_size=20)
```

```python
model.evaluate(x_test_sc, y_test)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```
