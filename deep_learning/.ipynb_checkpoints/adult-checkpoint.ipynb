```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./deep_data/adult.txt', header=None, names = ['age', 'workclass','fnlwgt', 'education', 'education_num', 
                                                                'marital_status', 'occupation', 'relationship', 'race'\
                                                                , 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 
                                                                'native-country', '50k'])
df.head()
```

```python
df['50k'].unique()
```

```python
df['target'] = df['50k'].apply(lambda x:1 if x == ' <=50K' else 0)
```

```python
df['target']
```

```python
df.info()
```

```python
y = df['target']
x = df.iloc[:,:14]
```

```python
x.info()
```

```python
x = pd.get_dummies(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=100, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(x_train)
x_train_sc = sc.transform(x_train)
x_test_sc = sc.transform(x_test)
```

```python
x
```

```python
model = Sequential()
model.add(Dense(1024, activation='relu', input_dim=108))
model.add(Dense(128, activation='relu'))
model.add(Dense(56, activation='relu'))
model.add(Dense(28, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'], optimizer='adagrad')
```

```python
hist = model.fit(x_train_sc, y_train, validation_split=0.2, epochs=50, batch_size=200)
```

```python
model.evaluate(x_test, y_test)
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train_sc, y_train)
pred = rfc.predict(x_test)
print(rfc.score(x_test, y_test))
```
