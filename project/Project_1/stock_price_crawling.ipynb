```python
df = pd.read_csv('data_3322_20220211.csv', encoding='euckr')
codes = df.단축코드.apply(lambda x:x).values
codes
```

```python
stock_code = []
for c in codes:
    if c[-1] not in [str(i) for i in range(10)]:
        code = f'{int(c[:-1]):05}'+ c[-1]
    else:
        code = f'{int(c):06}'
        
    stock_code.append(code)
    
stock_code
```

```python
for code in stock_code:
    print(code, '==========')
    
    aprile = [22, 47, 72, 96, 120, 145, 170, 194, 219, 243, 268]
    df = pd.DataFrame()

    for a in aprile:
        url = 'https://finance.naver.com/item/sise_day.naver?code={}&page={}'.format(code, a)
        driver.get(url)
        time.sleep(0.5)
        html = driver.page_source
        
        da = pd.read_html(html)[0]
        df = df.append(da, ignore_index=True)
        df.dropna(inplace=True)
        
        df['날짜'] = pd.to_datetime(df['날짜'], format='%Y-%m-%d')
        aprile_df = pd.DataFrame()
        
        aprile_df = df[df['날짜'].dt.day == 1]
        aprile_df = aprile_df.append(df[(df['날짜'].dt.year == 2012) & (df['날짜'].dt.day == 2)])
        aprile_df = aprile_df.append(df[(df['날짜'].dt.year == 2017) & (df['날짜'].dt.day == 3)])
        aprile_df = aprile_df.append(df[(df['날짜'].dt.year == 2018) & (df['날짜'].dt.day == 2)])
        
        aprile_df = aprile_df.sort_values('날짜')
        aprile_df = aprile_df.drop(['전일비', '시가', '고가', '저가'], axis=1)
        
        aprile_df.to_csv("./stock_price/{}.csv".format(code), mode='w')
```
