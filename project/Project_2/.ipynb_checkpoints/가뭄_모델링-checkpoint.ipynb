```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import datetime
```

※ SPI1, SPI1, SPI3,,,SPI24는 각각 1, 2, 3,,,24개월의 누적강수량 분석을 의미

※ 제공지점: 66개 지점 / 제공기간: 1991년 ~ / 발표주기: 일 1회

※ 지수범위별 가뭄단계 

   1.00 ≤ SPI : 습함

   0.99 ~ -0.99 : 정상

   -1.00 ~ -1.49 : 약한 가뭄(관심)

   -1.50 ~-1.99 : 보통 가뭄(주의)

   -2.00 ≥ SPI : 심한 가뭄(경계)

   -2.00 이하가 20일 이상 지속 : 극심한 가뭄(심각)

```python
f_li = ['가뭄지수_SPI_1991_2000','가뭄지수_SPI_2001_2010','가뭄지수_SPI_2011_2020','가뭄지수_SPI_2021_2021']
df_dry = pd.DataFrame()

for i in range(len(f_li)):
    df_1 = pd.read_csv('./{}.csv'.format(f_li[i]), encoding='euckr')
    df_dry = pd.concat([df_dry, df_1])

df_dry
```

```python
df_dry['일시'] = pd.to_datetime(df_dry['일시'], format='%Y-%m-%d')
```

```python
df_dry[df_dry['SPI24'].isna()]
```

```python
df_dry['월'] = df_dry['일시'].dt.month
```

```python
df_dry['일시'] = pd.Series(np.where(pd.notnull(df_dry['일시'])==True, df_dry['일시'], df_dry.groupby('월')['SPI24'].transform(np.mean)))
```
