```python
import requests
import pandas as pd
import numpy as np
```

```python
res = requests.get('https://poloniex.com/public?command=returnChartData&currencyPair=USDT_BTC&start=1405699200&end=9999999999&period=86400')
res.json
```

```python
df = pd.DataFrame(res.json())
df
```

```python
# close 값만 따로 value 뽑아낸거... 그 이름 죄송합니다. 
c_val = df['close'].values
```

```python
# 10일 간의 데이터를 가지고 하루 뒤 예측
window_size = 10 # 10일 간의 데이터 크기 설정

c_val
```

```python
c_val[0:10] # 10개, 이렇게 한묶음을 특징으로 쓰는거임
```

```python
# 그럼 앞에서부터 10개씩 모아야겠쥬?
c_val[0:10] # 이게 x[0]
c_val[1:11] # x[1]
c_val[2:12] # x[2]
# ...  이런식으로
```

```python
# 그래서 이 작업을 for문으로 돌려서 반복작업을 해주려고 하는거고 
for ind in range(len(c_val)):
    print(c_val[ind:ind+window_size])
# 이렇게 하면 에러가 뜨...ㄹ 거라고 생각했는데 안뜨네요..?
```

```python
# 마지막 부분을 보시면 원소가 하나만 들어갔어요. 네 없어서 안들어간겁니다. 
# 이게 사실 지혜누나가 헷갈리는 부분이었는데 
# 앞에서 부터 없어지는거다 아니다 뒤에꺼 없애는거다가 중요하지 않다고 말한게 앞에서부터 없애면 앞에서부터 없애는거고 뒤에서부터 없애면 뒤에서부터 없애는거라...
# 뭔말인지 나도 몰라요 암튼 
# 뒤에서부터 10번째까지는 10개씩 잘들어가요
len(c_val[-10:])
```

```python
# 근디 9번째부터는 다 안들어가요
len(c_val[-9:])
```

```python
#그래서 얘는 불완전한 데이터가 되는거라 버려야해요
# 그러면 window size크기 만큼 뒤에 있는 값은 가져갈수 있는거니깐
for ind in range(len(c_val)-window_size):
    print(c_val[ind:ind+window_size])
# 이렇게 해주면 마지막 값이 10개로 온전한 데이터가 되겠쥬?
len(c_val[ind:ind+window_size]) # 10이라고 나옵니다
```

```python
# 그래서 이걸 활용해서 x 데이터를 만들면 되는거지 
x = []
for ind in range(len(c_val)-window_size):
    x.append(c_val[ind:ind+window_size])
```

```python
# 그래서 여기선 x가 가져갈 범위를 1 더줄여줘야 마지막 y값까지 온전하게 가져갈 수 있답니다
x = []
y = []
for ind in range(len(c_val)-window_size):
    x.append(c_val[ind:ind+window_size])
    y.append(c_val[ind+window_size])
print(len(x),len(y))
```
