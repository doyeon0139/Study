```python
img = image.load_img('./image_test/hwayo_25/hwayo_25_1.jpg', target_size=(224, 224)) 

plt.imshow(img)
plt.show()
```

```python
x = image.img_to_array(img)
print('x_shape : ',x.shape)  # x_shape :  (150, 150, 3)
print('x_type : ',type(x))   # x_type :  <class 'numpy.ndarray'>
print(x.dtype)               # float32
# plt.imshow(x) : error 발생, 원인 : imshow는 0-1 float 또는 0-255 int 값을 input값으로 받음
plt.imshow(x)
plt.show()

x = x.reshape((1,) + x.shape)
```

```python
x.shape
```

```python
datagen = ImageDataGenerator(rotation_range = 90,
#                              width_shift_range=0.4,
#                              height_shift_range=0.4,
                             shear_range=0.2,
#                              zoom_range=0.2,
#                              zca_epsilon=1e-06,
                             brightness_range=(0,1),
#                              channel_shift_range=0.5,
                             fill_mode='constant',
                             cval=5,
#                              data_format='channels_last',
                             zca_whitening=True,
                             featurewise_std_normalization=True,
                             vertical_flip =True,
                             horizontal_flip =True)

idx = 0
fig = plt.figure(figsize=(10, 10))
axs = []
for batch in datagen.flow(x , batch_size=1): # 여기서 batch는 x가 됨
    axs.append(fig.add_subplot(5, 4, idx+1))
    axs[idx].imshow(image.array_to_img(batch[0]))
    idx += 1
    if idx%20 == 0:
        break
fig.tight_layout()
plt.show()
```

```python
import albumentations as A
```

```python
drink = cv2.imread("./image_test/hwayo_25/hwayo_25_1.jpg")
drink = cv2.cvtColor(drink, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(6,6))
plt.axis('off')
plt.imshow(drink)
```

```python
transform = A.Blur(blur_limit=(3,30), p=1) 
blur_drink = transform(image=drink)['image']
plt.figure(figsize=(6,6))
plt.axis('off')
plt.imshow(blur_drink)
```

```python
transform = A.GlassBlur(max_delta=3, iterations=2, p=1)
blur_drink = transform(image=drink)['image']
# plt.figure(figsize=(6,6))
plt.axis('off')
plt.imshow(blur_drink)
```

```python
axs = []

drink = cv2.imread("./image_test/hwayo_25/hwayo_25_1.jpg")
drink = cv2.cvtColor(drink, cv2.COLOR_BGR2RGB)
axs.append(drink)

transform = A.RGBShift() 
rgb_drink = transform(image=drink)['image']
axs.append(rgb_drink)

transform = A.HueSaturationValue() 
hue_drink = transform(image=drink)['image']
axs.append(hue_drink)

transform = A.ChannelShuffle() 
channel_drink = transform(image=drink)['image']
axs.append(channel_drink)

transform = A.CLAHE() 
clahe_drink = transform(image=drink)['image']
axs.append(clahe_drink)

transform = A.RandomContrast() 
rd_con_drink = transform(image=drink)['image']
axs.append(rd_con_drink)

transform = A.RandomGamma() 
rd_gam_drink = transform(image=drink)['image']
axs.append(rd_gam_drink)

transform = A.RandomBrightness() 
rd_bright_drink = transform(image=drink)['image']
axs.append(rd_bright_drink)

transform = A.Blur(blur_limit=(3,30), p=1) 
blur_drink = transform(image=drink)['image']
axs.append(blur_drink)

transform = A.MedianBlur() 
med_drink = transform(image=drink)['image']
axs.append(med_drink)

transform = A.ToGray() 
gray_drink = transform(image=drink)['image']
axs.append(gray_drink)

transform = A.JpegCompression() 
compre_drink = transform(image=drink)['image']
axs.append(compre_drink)
```

```python
fig = plt.figure(figsize=(13,13)) 
a = 1
 
for i in range(len(axs)):  
    ax = fig.add_subplot(3, 4, a)
    ax.imshow(axs[i])
    a+=1
    
plt.show()
```
