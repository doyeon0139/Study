```python
from summa.summarizer import summarize
import os
```

```python
open_path = 'E:/Study/project/Project_4/investing.com_news/'
fold_list = os.listdir(open_path)

for folder in fold_list:
    file_list = os.listdir(open_path+folder)
    os.mkdir('E:/Study/project/Project_4/dataset_summarization/'+folder)
    for file in file_list:
        data = pd.read_csv(open_path+folder+'/'+file, index_col=0)
        data.dropna(inplace=True)

        summary_li = []
        rat = 0.08

        for co in data.content:
            summa = summarize(co, rat)
            summary = summa.split('.')
            cnt = len(summary)
            while cnt < 3 and rat < 0.5:
                summa = summarize(co, rat)
                summary = summa.split('.')
                cnt = len(summary)
                rat += 0.01

            summary_li.append(summary)
    
        data_set = pd.DataFrame(data)
        data_set['content'] = summary_li
        
        save_path = 'E:/Study/project/Project_4/dataset_summarization/'+folder+'/'+file
        data_set.to_csv(save_path)
```
