```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import time, os, shutil, re
import requests
```

```python
url = 'https://www.huffpost.com/entry/starbucks-union-campaign-contract-fight_n_6259c871e4b066ecde1350f0'
res = requests.get(url)
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
```

```python
title = soup.select('#entry-header > header > div.top-header > h1')[0].text
```

```python
print(title)
```

```python
date = soup.select('#entry-header > header > div.bottom-header > div.bottom-header__left > div.timestamp > time:nth-child(1) > span')[0].text
```

```python
print(date)
```
