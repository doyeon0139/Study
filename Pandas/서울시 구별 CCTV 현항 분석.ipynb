### 서울시 구별 CCTV 현항 분석
* 구별 인구당 cctv 댓수 파악
* 인구당 cctv 수와 상관관계가 높은 요인은

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cc = pd.read_csv('./data/01. CCTV_in_Seoul.csv')
cc.head(3)
```

```python
cc.rename(columns = {'기관명' : '구별'}, inplace=True)
cc.head(3)
```

```python
pop = pd.read_excel('./data/01. population_in_seoul.xls', header=2, usecols = [1,3,6,9,13])
pop.head()
```

```python
pop.columns = ['구별','인구소','한국인','외국인','고령자']
pop.head(3)
```

```python
# CCTV 수를 기준으로 구를 정렬하시오
cc.sort_values('소계')
```

```python
# 2016년도에 cctv 수 증가율이 높은 상위 5개 구 출력
cc.sort_values('2016년', ascending=False).head(5)['구별'].values
```

```python
# 인구당 cctv 수가 가장 높은 구 상위 5개 출력
tt = pd.merge(cc,pop, left_on='구별',right_on='구별', how = 'inner')
tt
```

```python
tt['인C']=tt['소계']/tt['인구소']
tt
```

```python
tt.sort_values('인C', ascending=False).head(5)['구별'].values
```

```python
# 인구수, 외국인비율, 고령자 비율 중에서 인구당 cctv 수에 가장 큰 영향을 미치는 요소는 무엇인가
```

```python
sns.heatmap(tt.drop(['소계','2013년도 이전','2014년','2015년','2016년','한국인'], axis=1).corr(), cmap=plt.cm.rainbow)
```
