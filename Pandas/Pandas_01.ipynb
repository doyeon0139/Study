```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(np.random.randn(5,5), index=['a','b','c','d','e'], \
             columns=['one','two','three','four','five'])
```

```python
df.iloc[1:3, 1] = np.nan
df.iloc[2, 2:4] = np.nan
df
```

```python
df.isnull()
```

```python
# null값 합
df.isnull().sum()

df.isnull().sum().sum()
```

```python
# nan값이 잇는 row를 삭제
df.isnull().sum(axis=1)
```

```python
# nan값이 잇는 row를 삭제

df.dropna()
```

```python
# nan이 있는 column을 지우고 싶음
df.dropna(axis=1)
```

```python
# 전체 데이터가 nan인 경우만 삭제
df.dropna(how = 'all')
```

```python
# non을 가진 row의 데이터가 3개 이상이면 삭제 하지마
df.dropna(thresh = 3)
```

```python
# non 값을 0으로 채우기
df.fillna(0)
```

```python
# 앞에있는 값으로 non값을 채우라
df.fillna(method='ffill')
```

```python
# 뒤에있는 값으로 non값을 채우라
df.fillna(method='bfill')
```

```python
# column 평균값으로 채우기??????row??
df.fillna(df.mean())
```

```python
# ????? 평균채우기라는데 뭔 평균이라는지
df.interpolate()
```

```python
data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'Giant', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

ta = pd.DataFrame(data)
ta.head()
```

```python
# 팀별로 포인트의 평균값??????????
ta.groupby('Team')['Points'].mean()
```

```python
# Team별로 가장 좋은 ranking을 뽑는다
ta.groupby('Team')['Rank'].min()
```

```python
#  연도별 최고 점수는
ta.groupby('Year')['Points'].max()
```

```python
# 팀별 데이터 각각 몇개 인가?
ta.groupby('Team')['Points'].count()
```

```python
# 팀별, 연도별 포인트 최고점은?
ta.groupby(['Team','Year'])['Points'].max()
```
