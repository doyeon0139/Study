```python
items = {}

title_list = []
spec_list = []
price_list = []

for p in prod_item:
    try:
        title = p.select('div.prod_info > p > a')[0].text
        spec = p.select('div.prod_info > dl > dd > div')[0].text
        price = p.select('p.price_sect > a')[0].text
        
        print(title)
        print(spec)
        print(price)
    
        title_list.append(title)
        spec_list.append(spec)
        price_list.append(price)
        print('-----------------')
    except:
        pass
    
print(len(title_list))
print(len(spec_list))
print(len(price_list))
    
items['title'] = title_list
items['spec'] = spec_list
items['price'] = price_list

items
```

```python
item_df = pd.DataFrame(items)
item_df.head()
```

```python
drop_idx = []
for i, t in enumerate(item_df.title):
    if len(t.split()) < 2:
        drop_idx.append(i)
drop_idx
```

```python
item_df.drop(drop_idx[0], inplace=True)
```

```python
item_df
```

```python
comp_list = []
prd_nm_list = []

for p in item_df['title']:
    comp = p.split(' ', 1)[0]
    prd_nm = p.split(' ', 1)[1:][0]
    
    comp_list.append(comp)
    prd_nm_list.append(prd_nm)
```

```python
item_df['comp_nm'] = comp_list
item_df['prd_nm'] = prd_nm_list
```

```python
use_time = []
suction = []

for s in item_df.spec:
    for spec in s.split('/')[1:]:
        if '사용시간' in spec:
            time  =spec.split(':')[1]
        elif '흡입력' in spec:
            suc = spec.split(':')[1]
    use_time.append(time)
    suction.append(suc)
        
print(len(use_time))
print(len(suction))
```

```python
item_df['usetime'] = use_time
item_df['suction'] = suction
```

```python
item_df.drop(['title', 'spec'], axis=1, inplace=True)
```

```python
price_s = item_df['price'].str.replace('원', '')
price_s = price_s.str.replace(',', '')
item_df['price'] = price_s.apply(lambda x: int(x))
```

```python
def get_suction(value):
    value = value.upper()
    if 'AW' in value or 'W' in value:
        result = value.replace("A", "").replace('W', '')
        result = int(result.replace(',','').replace('~',''))
    elif 'PA' in value:
        result = value.replace('PA', '')
        result = int(result.replace(',','').replace('~',''))/100
    else:
        result = None
    return result
```

```python
item_df['suction'] = item_df['suction'].apply(get_suction)
```

```python
def convert_time_minute(time):
    if '시간' in time:
        hour = time.split('시간')[0]
        if '분' in time:
            minute = time.split('시간')[-1].split('분')[0]
        else:
            minute=0
    else:
        hour= 0
        minute = time.split('분')[0].split('~')[0]
    return int(hour)*60 + int(minute)
```

```python
item_df['usetime'] = item_df['usetime'].apply(convert_time_minute)
```
