```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pit
import seaborn as sns
```

```python
tt = pd.read_csv('./data/seoul_weather.csv', encoding='cp949', skiprows=11)
tt.head()
```

```python
tt.columns
```

```python
tt = tt.drop(['\t\t지점번호', '지점명', '\t최고기온시각', '최저기온시각'], axis=1)
```

```python
tt.columns = ['일시','평온','고온','저온']
tt.head()
```

```python
tt.info
```

```python
tt['일시'] = pd.to_datetime(tt['일시'])
tt.info()
```

```python
tt['일시'].dt.dayofweek
```

```python
# 8월의 최고기온을 추출해서 라인그래프로 그리시오
```

```python
tt.loc[tt['일시'].dt.month==8]['고온'].plot()
```

```python
tt.loc[tt['일시'].dt.month==8]['고온'].plot(kind='hist', bins=30)
```

```python
# 매년 8월 중에 최고기온 하나를 뽑아서 라인 차트로 그리시오
mon = pd.DataFrame()
mon['고온'] = tt[tt['일시'].dt.month == 8]['고온']
mon
```

```python
mon['year'] = tt['일시'].dt.year
mon
```

```python
mon.groupby('year')['고온'].max().plot()
```
