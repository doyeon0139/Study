```python
theme_num = [1005,1006,1007,1008,1009,1010,1011,1012,1013,1014,1015,1016,1017,1018,1019,1020,1021,1022,1024,1025,1026,1027 ]
```

```python
theme_name = ['음식료품','섬유의복','종이목재','화학','의약품','비금속광물','철강금속',
 '기계','전기전자','의료정밀','운수장비','유통업','전기가스업','건설업','운수창고업','통신업',
 '금융업','은행','증권','보험','서비스업','제조업']
```

```python
for th in range(len(theme_num))[1:]:
    pdf_li = stock.get_index_portfolio_deposit_file(str(theme_num[th]))
    df = pd.DataFrame({'stock_code':pdf_li})
    df['stock_code'] = df['stock_code'].astype('int64')
    theme_df = pd.merge(df, stock_df, how='left', on='stock_code')
    theme_df.to_csv(f'C:/Users/bitcamp/Desktop/final_data/테마별/{theme_name[th]}.csv')
```
